```python
from pathlib import Path
import pickle

folder_path = Path('/content/drive/MyDrive/final-year-project/stories-pkl-file')
all_data = []

for file in folder_path.glob('*.pkl'):
    with open(file, 'rb') as f:
        data = pickle.load(f)
        all_data.append(data)
```

```python
type(all_data)
```

```python
all_sentences = []
for data in all_data:
    for sentence in data:
        all_sentences.append(sentence)
```

```python
from bnlp import CRFTaggerTrainer
trainer = CRFTaggerTrainer()
model_name = "ner_model.pkl"
```

```python
train_data = all_sentences
test_data = all_sentences
trainer.train(model_name, train_data, test_data)
```
